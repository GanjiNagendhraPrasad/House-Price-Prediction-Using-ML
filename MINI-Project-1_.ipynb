#train_test_split:


```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train),len(y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
X=x_train.values
X.shape[1]
```

```python
y_train.values.shape
```

```python
# Convert to numpy
X = x_train.values
y = y_train.values

n_features = X.shape[1]

m = []   # to store m1, m2, ..., m18

y_mean = np.mean(y)

for j in range(n_features):
    xj = X[:, j]
    xj_mean = np.mean(xj)

    numerator = np.sum((xj - xj_mean)**2 * (y - y_mean)**2)
    denominator = np.sum((xj - xj_mean) ** 2)

    mj = numerator / denominator
    m.append(mj)

m = np.array(m)
m
```

```python
X_mean = np.mean(X, axis=0)   # mean of each column
c = y_mean - np.sum(m * X_mean)
c
```

```python
X_mean.shape
```

```python
zero_variance_cols = []
for j in range(n_features):
    if np.var(X[:, j]) == 0:
        zero_variance_cols.append(j)
zero_variance_cols
```

```python
# Find non-zero variance columns
valid_cols = []
for j in range(n_features):
    if np.var(X[:, j]) != 0:  #Variance tells us whether a column has changing values or not.
        valid_cols.append(j)
X_valid = X[:, valid_cols]
```

```python
valid_cols
```

```python
X_valid.shape
```

```python
m = []
y_mean = np.mean(y)

for j in range(X_valid.shape[1]):
    xj = X_valid[:, j]
    xj_mean = np.mean(xj)

    numerator = np.sum((xj - xj_mean)**2 * (y - y_mean)**2)
    denominator = np.sum((xj - xj_mean) ** 2)

    mj = numerator / denominator
    m.append(mj)

m = np.array(m)
m
```

```python
X_mean = np.mean(X_valid, axis=0)

c = y_mean - np.sum(m * X_mean)
print(c)
```

#train data

```python
X_train_valid = x_train.values[:, valid_cols]

y_train_pred = np.dot(X_train_valid, m) + c #(np.dot)-Multiply corresponding values and add them. [np.dot(x, m) = m1*x1 + m2*x2]
```

```python
len(y_train_pred)
```

```python
print(X_train_valid.shape)
print(m.shape)
print(y_train_pred.shape)
```

```python
training_data = pd.DataFrame({'y_train':y_train.values.flatten(),'ans_train':y_train_pred })
training_data
```

```python
# mean square error
s = 0
for i in training_data.index:
  s  = s + (training_data['y_train'][i] - training_data['ans_train'][i])**2

print(f'Loss value from MSE : {s / len(training_data)}')
```

```python
# root mean square error
s = 0
for i in training_data.index:
  s  = s + (training_data['y_train'][i] - training_data['ans_train'][i])**2

print(f'Loss value from RMSE : {np.sqrt(s / len(training_data))}')
```

```python
# Absolute mean square error
s = 0
for i in training_data.index:
    s = s + abs(training_data['y_train'][i] - training_data['ans_train'][i])

print(f'Loss value from AMSE : {s / len(training_data)}')
```

```python
# Train Accuracy

numerator = 0
denominator = 0
for i in training_data.index:
  numerator  = numerator + (training_data['y_train'][i] - training_data['ans_train'][i])**2
  denominator = denominator + (training_data['y_train'][i] - training_data['y_train'].mean())**2


print(f'Train Accuracy By r2_score formula : {1 - (numerator) / denominator}')
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,r2_score
mean_squared_error(y_train,y_train_pred)
```

```python
root_mean_squared_error(y_train,y_train_pred)
```

```python
mean_absolute_error(y_train,y_train_pred)
```

```python
r2_score(y_train,y_train_pred)
```
